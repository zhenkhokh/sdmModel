```python
def plotSpectr(df,legend):
    power = np.square(np.abs(fftpack.fft(np.blackman(df.size/2)))).sum()
    line, = plt.semilogx(20*np.log10(np.abs(fftpack.fft(df.output*np.blackman(df.size/2))))-10*np.log10(power))
    plt.ylabel('db')
    plt.xlabel('Hz, if obeservation time is 1 sec')
    plt.title('Sigma-delta spectr')
    line.set_label(legend)
    plt.legend()
    
```
